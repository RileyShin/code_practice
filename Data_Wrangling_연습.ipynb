 ## ✅ 1차 탐색결과
    

1. customer_id : 


> - 자료형 타입이 실수형태임, 명목형 자료임으로 실수형에서 
정수형으로 바꾼후 문자형 타입으로 변경 필요
- 0번째 객체의 값을 100번째 인덱스로 변경

2.   email 


> - 결측값이 존재함 -대체할 방법이 없으므로 삭제를 하는 방향으로 진행
- '@'가 없는 개체가 있으므로 '@'가 들어있지 않는 행은 넣기
- 이메일의 앞만 있다면, 'sakilacustomer.org'를 넣기

3. address_id


> - 수치형 형태이기 때문에 customer_id에 4가 더해진 형태인듯 함 
- 자료형타입이 실수형이므로, '실수형 → 정수형 → 문자형'으로 바꾸기

4. name


> 문자들 가운데 , , -,을 공백으로 바꾸고 시작점들을 capitalize()를 사용해 대문자화 해 두개로 쪼개 성, 이름 변수에 저장

5. age


> 뒤에 두개를 추출한 후, 타입을 변경하기











### ❤ customer_id 데이터 탐색 & 정제

```python
diss.query('customer_id > 100')
```

```python
diss.customer_id.loc[0]=100.0
```

```python
# customer_id 기준 정렬
diss.sort_values('customer_id').head()
```

```python
# 타입 변경하기 (실수 → 정수)
diss['customer_id']=diss.customer_id.astype('int')
diss.head()
```

```python
# customer_id를 기준으로 정렬하기
diss=diss.sort_values('customer_id')
diss.tail()
```

```python
# 타입 변경하기 (정수 → 문자)
diss['customer_id']=diss.customer_id.astype('str')
diss.head()
```

### ❤ email 데이터 탐색 및 정제

```python
# 데이터 복제하기
diss_cp=diss.copy()
```

```python
# 본 email 형식에 맞지 않는 것 확인
# 결측치확인

diss_cp[~diss_cp['email'].str.contains('@',na=False)]
```

```python
# email 형식에 맞게 변경하기 
diss_cp['email']=diss_cp.email.str.replace(',','@')
```

```python
diss_cp[~diss_cp['email'].str.contains('@',na=False)]
```

```python
diss_cp.email.loc[96]=diss_cp.email.loc[96]+'@sakilacustomer.org'
```

```python
diss_cp.email.loc[96]
```

```python
diss=diss_cp
```

```python
#email의 결측값 채우기
name=diss_cp['name'].str.split(',')

diss_cp['email']=diss_cp['email'].fillna(name.str.get(0).str.lower()+'.'+name.str.get(1).str.lower()+'@sakilacustomer.org')
```

```python
#email에 공백이 있는 값들 찾기
diss_cp['email']=diss_cp['email'].str.replace(' ','')
diss_cp['email']
```

```python
# 원본 데이터 리뉴얼!
diss=diss_cp
```

### ❤ address_id 데이터 탐색 및 정제

```python
# customer_id의 타입을 잠시 정수형으로 변경하기
diss_cp['customer_id']=diss_cp['customer_id'].astype('int')
```

```python
diss_cp['address_id']=diss_cp['customer_id']+4
```

```python
diss_cp['address_id'].head()
```

```python
## customer_id, address_id의 타입을 다시 object로 변경하기

diss_cp['customer_id']=diss_cp['customer_id'].astype('str')
diss_cp['address_id']=diss_cp['address_id'].astype('str')
diss_cp
```

### ❤ name 데이터 탐색 및 정제

```python
diss_cp['name']=diss['name'].str.replace('-','')
```

```python
diss_cp['name']=diss['name'].str.replace(',','')
```

```python
diss_cp['name']
```

```python
name_split=diss_cp['name'].str.split(" ")

diss_cp['first_name']=name_split.str.get(0)
diss_cp['last_name']=name_split.str.get(1)
```

```python
diss_cp.head()
```

```python
# nan있는 행 찾기
diss_cp[diss_cp['first_name'].isna() | diss_cp['last_name'].isna()]
```

```python
#이메일에서 이름 채우기 (nan값 해결하기)
email=diss_cp['email'].str.split('[.,@]')
```

```python
diss_cp['first_name']=email.str.get(0)
diss_cp['last_name']=email.str.get(1)
```

```python
# nan있는 행 찾기
diss_cp[diss_cp['first_name'].isna() | diss_cp['last_name'].isna()]
```

```python
name=diss_cp[diss_cp['name'].isnull() | diss_cp['email'].isnull()]
name
```

```python
diss_cp['name']=diss_cp['name'].fillna(diss_cp['first_name'].str.capitalize()+' '+diss_cp['last_name'].str.capitalize())
```

```python
diss_cp.head()
```

```python
diss_cp['name'].loc[2]=diss_cp['first_name'].loc[2].capitalize()+' '+diss_cp['last_name'].loc[2].capitalize()
diss_cp['name'].loc[2]
```

```python
# 대문자화되지 않은 값 확인
diss_cp[~diss_cp['name'].str.contains('[A-Z]')]
```

```python
diss_cp['name'].loc[11]=diss_cp['first_name'].loc[11].capitalize()+' '+diss_cp['last_name'].loc[11].capitalize()
diss_cp['name'].loc[11]
```

```python
diss_cp
```

```python
diss_cp=diss_cp.drop(['first_name','last_name'],axis=1)
```

### ♥ age 데이터 탐색 및 정제

```python
diss_cp.info()
```

```python
diss_cp['age']=diss_cp['age'].str[5:]
diss_cp['age'].head()
```

```python
#age에서 결측값이나 숫자가 아닌 데이터 찾기
diss_cp[diss_cp['age'].isna()|diss_cp['age'].str.contains('[^0-9]')] 
```

```python
#age의 타입 변경하기 
diss_cp['age']=diss_cp['age'].astype('int')
diss_cp.info()
```
